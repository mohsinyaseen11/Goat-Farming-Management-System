```python
import pandas as pd
goats=pd.read_csv("goats.csv")
field=["ID","Name","Breed","Gender","DOB/P. D","PurchasePrice","HealthStatus"]


def add_goats(goats,field):
    datalist=[]
    for i in field:
        value=input(f"Enter {i}: ")
        datalist.append(value)
    goats.loc[len(goats)]=datalist
    goats.to_csv("goats.csv", index=False)


def show_goats(goats):
    print(goats)



def update_goats(goats,field):
    old=int(input("Please Enter ID"))
    for i in range(len(field)):
        print(i,field[i])

    s=int(input("Select any One"))
    if s==0:
        change=int(input("Enter New ID"))
        goats.loc[goats["ID"]==old,"ID"]=change

    elif s==1:
        change=input("Enter New Name")
        goats.loc[goats["ID"]==old,"Name"]=change

    elif s==2:
        change=input("Enter New Breed Name")
        goats.loc[goats["ID"]==old,"Breed"]=change


    elif s==3:
        change=input("Enter New Gender")
        goats.loc[goats["ID"]==old,"Gender"]=change



    elif s==4:
        change=input("Enter New DOB/P. D")
        goats.loc[goats["ID"]==old,"DOB/P. D"]=change



    elif s==5:
        change=input("Enter New PurchasePrice")
        goats.loc[goats["ID"]==old,"PurchasePrice"]=change

    elif s==6:
        change=input("Enter New HealthStatus")
        goats.loc[goats["ID"]==old,"HealthStatus"]=change


    goats.to_csv("goats.csv", index=False)
    return goats




def delete_goat(goats):
    old=int(input("Please Enter ID"))
    goats=goats[goats["ID"]!=old]       
    goats.to_csv("goats.csv", index=False)
    return goats



def search_goat(goats):
    old=int(input("Please Enter ID"))
    print(goats.loc[goats["ID"]==old])
    
         


    

option=1
print("Welcome To Our Goat Farming Management System")
while option==1:
    print("1.Add New Goat\n","2.View All Goats\n","3.Update Goat Info\n","4.Delete Goat by ID\n","5.Search Goat by ID\n","6.Exit")
    select=int(input("Select Any One Option"))
    if select==1:
        goats=add_goats(goats,field)


    elif select==2:
        show_goats(goats)


    elif select==3:
        update_goats(goats,field)




    
    elif select==4:
         delete_goat(goats)



    elif select==5:
        search_goat(goats)



    elif select==6:
        option=0
        
    else:
        print("Please Enter Valid Option")
    
    
```
